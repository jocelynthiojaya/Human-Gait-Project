```python
import pandas as pd
data = pd.read_csv('6_Janewalk.csv')
data.head()
```

```python

context_columns = ['id','gender','age']
sequence_index = 'frame'
from sdv.timeseries import PAR

model = PAR(
#entity_columns=entity_columns,
#context_columns=context_columns,
sequence_index=sequence_index,
)

model.fit(data)
```

```python
synthetic_data = model.sample()
synthetic_data.head()
```

```python
from sdv.evaluation import evaluate

evaluate(synthetic_data, data)
```

```python
evaluate(synthetic_data, data, aggregate=False)
```

```python
new_data = pd.concat([data, synthetic_data])
print(len(new_data.index))
```

```python
new_data.to_csv('9_RaymondwalkSynthetic.csv', index=False)
```
